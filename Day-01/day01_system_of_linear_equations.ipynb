Solving a System using numpy.linalg.solve

```python
import numpy as np

# Description: Solving 2 linear equations with 2 unknowns
# Example:
# 2x + 3y = 5
# 4x -  y = 1

# Coefficient matrix A
A = np.array([[2, 3],
              [4, -1]])

# Constant matrix b
b = np.array([5, 1])

# Solving for x (variables)
solution = np.linalg.solve(A, b)

print("Solution [x, y]:", solution)
```

Checking if a system is singular (no solution or infinite solutions)

```python
# Description: Checking determinant to know if system is singular

det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

if det_A == 0:
    print("System is Singular (No unique solution)")
else:
    print("System is Non-singular (Unique solution exists)")
```

Visualizing the Equations (Optional for Better Understanding)

```python
import matplotlib.pyplot as plt

# Description: Plotting the two equations as lines

# Define the range of x
x = np.linspace(-10, 10, 100)

# Define the two equations
y1 = (5 - 2*x)/3
y2 = (4*x - 1)

# Plot
plt.plot(x, y1, label='2x + 3y = 5')
plt.plot(x, y2, label='4x - y = 1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Linear Equations')
plt.legend()
plt.grid(True)
plt.show()
```
